```python
import matplotlib.pyplot as plt                                  #For inline image showing
```

# Load Cascade Classifier

```python
face_cascade_file = 'Cascade Classifier/face-detect.xml'
face_cascade = cv2.CascadeClassifier(face_cascade_file)
```

# Face Detection

## Face Detection on Still Image 

- Load image
- Resize untuk mempercepat proses
- Convert ke RGB dan GRAY
- Pencarian wajah dengan cascade classifiernya
- Looping untuk memberi penanda pada wajah
- Show gambarnya

```python
image_path = 'sample image/image from freepik.jpg'
img = cv2.imread(image_path)                                       # Read image
img = cv2.resize(img, (640, 360), interpolation=cv2.INTER_LINEAR)  # Resize image for the sake of time

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     # Convert ke RGB
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)                   # Convert ke GRAY (hitam putih)

faces = face_cascade.detectMultiScale(gray_img, 1.3, 0)            # Proses pencarian wajah 
                                                                   # <cascade_file>.detectMultiScale(<frame>, <scale_factor>, <min_neighbors>)
for x, y, w, h in faces:                                           # Looping untuk wajah yang terdeteksi
    cv2.rectangle(rgb_img, (x, y), (x+w, y+h), (0, 255, 0), 3)     # Gambar Box untuk setiap wajah yang terdeteksi


plt.imshow(rgb_img)
plt.show()
```

## Face Detection on Video Stream

- Connect camera
- Looping untuk membaca seluruh frame camera
- Convert ke GRAY
- Pencarian wajah dengan cascade classifiernya
- Looping untuk memberi penanda pada wajah
- Show gambarnya
- Key handler untuk exit program
- Putuskan koneksi camera dan tutup jendela

```python
cam = cv2.VideoCapture(0)                                           # Akses Kamera
while True:
    ret, frame = cam.read()                                         # Membaca setiap frame dari stream kamera 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)                  # Mengubah mode BGR ke GRAY (hitam putih)
    
                                                                    # Proses pencarian wajah 
    faces = face_cascade.detectMultiScale(gray, 1.3, 3)             # <cascade_file>.detectMultiScale(<frame>, <scale_factor>, <min_neighbors>)
    for x, y, w, h in faces:                                        # Looping semua wajah yang terdeteksi
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 3)    # Gambar box untuk setiap wajah
    
    cv2.imshow('Face Detect Video', frame)                          # Jendela untuk menampilkan hasil
    
    if cv2.waitKey(1) & 0xff == ord('x'):                           # Exit dengan tombol x
        break

cam.release()                                                       # Menyudahi akses kamera
cv2.destroyAllWindows()                                             # Menutup jendela
```
